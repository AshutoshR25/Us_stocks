```python
ticker = 'AAPL'
multiplier = '1'
timespan = 'minute'
start = '2023-03-24'        
end = '2023-03-24'
limit = '50000'
API = "RSv22agUPFHpC_KyOijn6aw0uY39z5tV"

url = f'https://api.polygon.io/v2/aggs/ticker/{ticker}/range/{multiplier}/{timespan}/{start}/{end}?limit={limit}&apiKey={API}'

url
```

```python
import requests
import pandas as pd

data = requests.get(url).json()

data
```

```python
df = pd.DataFrame(data['results'])

df['date'] = pd.to_datetime(df['t'], unit='ms')

df
df.to_csv(ticker+('.csv'), header=None)
```
